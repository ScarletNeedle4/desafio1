# Análisis de Datos de Ventas - Challenge 1 Alura Latam

Este notebook contiene el análisis completo de los datos de ventas de cuatro tiendas diferentes. El objetivo es responder a las preguntas de negocio planteadas en el challenge.

## Importación de librerías y datos

```python
# Importar las librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt

# Configuración para gráficos
plt.rcParams['axes.unicode_minus'] = False

# URLs de los archivos de datos
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
]

# Cargar cada tienda y añadir una columna para identificarla
tiendas = []
for i, url in enumerate(urls, start=1):
    df_tienda = pd.read_csv(url)
    df_tienda['Tienda'] = f'Tienda {i}'
    tiendas.append(df_tienda)

# Unir todos los DataFrames en uno solo
df = pd.concat(tiendas, ignore_index=True)

# Crear columna con el valor total de la venta (Precio + Costo de envío)
df['Valor Total Venta'] = df['Precio'] + df['Costo de envío']

# Mostrar las primeras filas para verificar la carga
print("Vista previa del dataframe consolidado:")
print(f"Total de registros: {len(df)}")
df.head()
```

## 1. Análisis de facturación

Calculamos la facturación total (suma de precio + envío) por tienda para identificar la de mayor rendimiento.

```python
# Agrupar por tienda y sumar el valor total de las ventas
facturacion_por_tienda = df.groupby('Tienda')['Valor Total Venta'].sum().reset_index()
facturacion_por_tienda = facturacion_por_tienda.sort_values(by='Valor Total Venta', ascending=False)

# Mostrar resultados
print("=" * 50)
print("FACTURACIÓN TOTAL POR TIENDA")
print("=" * 50)
for index, row in facturacion_por_tienda.iterrows():
    print(f"{row['Tienda']}: ${row['Valor Total Venta']:,.0f}")

# Gráfico de facturación
plt.figure(figsize=(8, 5))
barras = plt.bar(facturacion_por_tienda['Tienda'], facturacion_por_tienda['Valor Total Venta'], 
                color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
plt.title('Facturación Total por Tienda', fontsize=14)
plt.xlabel('Tienda')
plt.ylabel('Facturación Total ($)')

# Añadir etiquetas con los valores
for barra in barras:
    yval = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, yval, 
             f'${yval/1e6:.1f}M', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()

# Análisis
tienda_top = facturacion_por_tienda.iloc[0]['Tienda']
facturacion_top = facturacion_por_tienda.iloc[0]['Valor Total Venta']
print(f"\n✅ La tienda con mayor facturación es {tienda_top} con ${facturacion_top:,.0f}")
```

## 2. Ventas por categoría

Analizamos la facturación por categoría de producto para entender qué tipos de productos generan más ingresos.

```python
# Agrupar por categoría y sumar el valor total
ventas_por_categoria = df.groupby('Categoría del Producto')['Valor Total Venta'].sum().reset_index()
ventas_por_categoria = ventas_por_categoria.sort_values(by='Valor Total Venta', ascending=False)

# Mostrar resultados
print("=" * 50)
print("FACTURACIÓN POR CATEGORÍA DE PRODUCTO")
print("=" * 50)
for index, row in ventas_por_categoria.iterrows():
    porcentaje = (row['Valor Total Venta'] / df['Valor Total Venta'].sum()) * 100
    print(f"{row['Categoría del Producto']}: ${row['Valor Total Venta']:,.0f} ({porcentaje:.1f}%)")

# Gráfico de ventas por categoría
plt.figure(figsize=(10, 6))
barras = plt.bar(ventas_por_categoria['Categoría del Producto'], ventas_por_categoria['Valor Total Venta'], 
                color='skyblue')
plt.title('Facturación Total por Categoría de Producto', fontsize=14)
plt.xlabel('Categoría')
plt.ylabel('Facturación Total ($)')
plt.xticks(rotation=45, ha='right')

# Añadir etiquetas
for barra in barras:
    yval = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, yval, 
             f'${yval/1e6:.1f}M', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

categoria_top = ventas_por_categoria.iloc[0]['Categoría del Producto']
print(f"\n✅ La categoría con mayor facturación es: {categoria_top}")
```

## 3. Calificación promedio de la tienda

Calculamos la calificación promedio (1-5) para cada tienda.

```python
# Calcular calificación promedio por tienda
calificacion_promedio = df.groupby('Tienda')['Calificación'].mean().reset_index()
calificacion_promedio = calificacion_promedio.sort_values(by='Calificación', ascending=False)

# Mostrar resultados
print("=" * 50)
print("CALIFICACIÓN PROMEDIO POR TIENDA")
print("=" * 50)
for index, row in calificacion_promedio.iterrows():
    print(f"{row['Tienda']}: {row['Calificación']:.2f} / 5.00")

# Gráfico de calificaciones
plt.figure(figsize=(8, 5))
barras = plt.bar(calificacion_promedio['Tienda'], calificacion_promedio['Calificación'],
                color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
plt.title('Calificación Promedio por Tienda', fontsize=14)
plt.xlabel('Tienda')
plt.ylabel('Calificación Promedio')
plt.ylim(0, 5.5)

# Añadir etiquetas
for barra in barras:
    yval = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, yval + 0.1, 
             f'{yval:.2f}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()

tienda_mejor_calificacion = calificacion_promedio.iloc[0]['Tienda']
print(f"\n✅ La tienda con mejor calificación promedio es: {tienda_mejor_calificacion}")
```

## 4. Productos más y menos vendidos

Identificamos los productos con mayor y menor frecuencia de venta.

```python
# Contar la frecuencia de cada producto
ventas_por_producto = df['Producto'].value_counts().reset_index()
ventas_por_producto.columns = ['Producto', 'Cantidad de Ventas']

print("=" * 50)
print("TOP 10 PRODUCTOS MÁS VENDIDOS")
print("=" * 50)
top10 = ventas_por_producto.head(10)
for index, row in top10.iterrows():
    print(f"{index+1}. {row['Producto']}: {row['Cantidad de Ventas']} ventas")

# Gráfico de productos más vendidos
plt.figure(figsize=(10, 5))
barras = plt.barh(top10['Producto'][::-1], top10['Cantidad de Ventas'][::-1], color='lightcoral')
plt.title('Top 10 Productos Más Vendidos', fontsize=14)
plt.xlabel('Cantidad de Ventas')
plt.ylabel('Producto')
for barra in barras:
    ancho = barra.get_width()
    plt.text(ancho + 0.5, barra.get_y() + barra.get_height()/2, 
             f'{int(ancho)}', ha='left', va='center', fontsize=9)
plt.tight_layout()
plt.show()

print("\n" + "=" * 50)
print("TOP 10 PRODUCTOS MENOS VENDIDOS")
print("=" * 50)
bottom10 = ventas_por_producto.tail(10)
for index, row in bottom10.iterrows():
    print(f"{index+1}. {row['Producto']}: {row['Cantidad de Ventas']} ventas")

# Gráfico de productos menos vendidos
plt.figure(figsize=(10, 5))
barras = plt.barh(bottom10['Producto'][::-1], bottom10['Cantidad de Ventas'][::-1], color='lightblue')
plt.title('Top 10 Productos Menos Vendidos', fontsize=14)
plt.xlabel('Cantidad de Ventas')
plt.ylabel('Producto')
for barra in barras:
    ancho = barra.get_width()
    plt.text(ancho + 0.1, barra.get_y() + barra.get_height()/2, 
             f'{int(ancho)}', ha='left', va='center', fontsize=9)
plt.tight_layout()
plt.show()

producto_top = ventas_por_producto.iloc[0]['Producto']
print(f"\n✅ El producto más vendido es: {producto_top}")
```

## 5. Envío promedio por tienda

Calculamos el costo de envío promedio por transacción para cada tienda.

```python
# Calcular costo de envío promedio por tienda
envio_promedio = df.groupby('Tienda')['Costo de envío'].mean().reset_index()
envio_promedio = envio_promedio.sort_values(by='Costo de envío', ascending=False)

# Mostrar resultados
print("=" * 50)
print("COSTO DE ENVÍO PROMEDIO POR TIENDA")
print("=" * 50)
for index, row in envio_promedio.iterrows():
    print(f"{row['Tienda']}: ${row['Costo de envío']:,.0f}")

# Gráfico de envío promedio
plt.figure(figsize=(8, 5))
barras = plt.bar(envio_promedio['Tienda'], envio_promedio['Costo de envío'],
                color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
plt.title('Costo de Envío Promedio por Tienda', fontsize=14)
plt.xlabel('Tienda')
plt.ylabel('Costo de Envío Promedio ($)')

# Añadir etiquetas
for barra in barras:
    yval = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, yval + 200, 
             f'${yval:,.0f}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()

tienda_mayor_envio = envio_promedio.iloc[0]['Tienda']
print(f"\n✅ La tienda con mayor costo de envío promedio es: {tienda_mayor_envio}")
```
